Accessing Characters â€” Indexing

```python
#Positive Index
a = "python"

print(a[0])  # p
print(a[1])  # y
print(a[5])  # n
```

```python
#Negative Index
print(a[-1])  # n
print(a[-2])  # o
print(a[-6])  # p
```

String Slicing

```python
# string[start : end : step]
```

```python
a = "python"

print(a[0:3])   # pyt
print(a[2:5] )  # tho
print(a[:4]   ) # pyth
print(a[3:]    )# hon
print(a[:]     )# python
```

Step Slicing

```python
print(
a[::2] +'\n'+   # pto
a[::-1]  # nohtyp (reverse)
)
```

Traversing Strings

```python
for i in range(len(a)):
    print(a[i])
```

```python
for ch in "python":
    print(ch)
```

Immutability & Modification Trick

```python
a = "python"
a = "P" + a[1:]
a
```

Counting Characters

```python
# Count specific char

a = "banana"
a.count("a")   # 3
```

```python
# Count vowels

a = "programming"
vowels = "aeiou"

count = 0
for ch in a:
    if ch in vowels:
        count += 1

print(count)
```

```python
# Interview shortcut

sum(1 for ch in a if ch in "aeiou")
```

```python
# Finding Alternative Alphabets

ch = "a"

next_ch = chr(ord(ch) + 1)
print(next_ch)   # b



# Shift String

a = "abc"

shifted = ""
for ch in a:
    shifted += chr(ord(ch) + 1)

print(shifted)   # bcd
```

Alternate Characters

```python
a = "python"

print(a[::2])   # pto

#start from 2nd
print(a[1::2])   # pto
```

Case Manipulation

```python
a = "PyThOn"

print(
a.lower(),   # python
a.upper(),   # PYTHON
a.title(),   # Python
a.capitalize(),  # Python
a.swapcase()  )  # pYtHoN
```

Remove Vowels

```python
a = "programming"

result = ""
vowel = "aeiou"

for i in a:
    if i not in vowel:
        result +=  i

print(result) 
```

Frequency Count

```python
a = "banana"

freq = {}
for ch in a:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```
