Local Scope: Variable inside a function

```python
def func():
    x = 10
    print(x)

func()
```

Global Scope:Defined outside all functions.

```python
x = 100

def func():
    print(x)

func()
```

```python
x = 100

def func():
    x = 50   # New local variable

func()
print(x)   # 100
```

global Keyword:Used to modify global variable inside function.

```python
x = 100

def func():
    global x
    x = 50

func()
print(x)   # 50
```

⚠️ Interview note:

Avoid excessive global use → bad design.


Enclosing Scope (Nested Functions): Also called non-local scope.

```python
def outer():
    x = 10

    def inner():
        print(x)

    inner()

outer()
```

nonlocal Keyword:Used to modify enclosing variable.

Without nonlocal

```python
def outer():
    x = 10

    def inner():
        x = 20   # New local
        print(x)

    inner()
    print(x)

outer()
```

With nonlocal

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x = 20

    inner()
    print(x)

outer()
```

LEGB Resolution — Step by Step

Example:

```python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        x = "local"
        print(x)

    inner()

outer()
```

```python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        print(x)

    inner()

outer()
```

```python
x = 10

def func():
    print(x)
    x = 20

func()
```

```python
x = 5

def func():
    global x
    x += 1

func()
print(x)
```

```python
x = 5

def func():
    x += 1

func()
print(x)
```
