Typecasting (Type Conversion) — VERY IMPORTANT

Typecasting = converting one data type → another.

Two types:

1. Implicit

2. Explicit

Implicit Typecasting

```python
a = 10      # int
b = 2.5     # float

c = a + b
print(c)
print(type(c))
```

Explicit Typecasting

```python
x = "10.2"
y = int(float(x))

print(y)
print(type(y))
```

Input + Typecasting Together

```python
age = input("Enter age: ")
print(type(age))   # str
```
