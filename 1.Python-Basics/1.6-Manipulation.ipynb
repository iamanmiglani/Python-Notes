Manipulation

Swapping Variables

```python
a = 10
b = 20

a, b = b, a
print(a, b)
```

```python
temp = a
a = b
b = temp
print(a, b)
```

String Manipulation

```python
first = "Aman"
last = "Miglani"

full_name = first + " " + last
print(full_name)
```

```python
salary = 50000
salary = salary + 5000

print(salary)
```

```python
count = 0

count += 1
count += 1

print(count)   # 2
```

```python
x = 10
y = x
x = 20

print(y)
```
