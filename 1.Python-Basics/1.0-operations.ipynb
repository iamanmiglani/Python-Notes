3Ô∏è‚É£ Arithmetic Operators

Used for mathematical calculations.

| Operator | Meaning        | Example |
| -------- | -------------- | ------- |
| +        | Add            | 5+2     |
| -        | Subtract       | 5-2     |
| *        | Multiply       | 5*2     |
| /        | Division       | 5/2     |
| //       | Floor division | 5//2    |
| %        | Modulus        | 5%2     |
| **       | Power          | 5**2    |


```python
5//2,5%2,5**2
```

```python
a = 7
b = 2

print(a / b)   # 3.5
print(a // b)  # 3
print(a % b)   # 1
print(a ** b)  # 49
```

#### Edge cases

```python
-7 // 2   # -4  (floors, not truncates)
```

5Ô∏è‚É£ Logical Operators

Used to combine conditions.

| Operator | Meaning   |
| -------- | --------- |
| and      | Both true |
| or       | Any true  |
| not      | Reverse   |


```python
x = 10

print(x > 5 and x < 20)  # True
print(x < 5 or x == 10)  # True
print(not(x > 5))        # False
```

Short-Circuiting

```python
False and print("Hi")   # Won‚Äôt run print
True or print("Hi")     # Won‚Äôt run print
```

7Ô∏è‚É£ Bitwise Operators

Operate on binary bits.

| Operator | Meaning     |    |
| -------- | ----------- | -- |
| &        | AND         |    |
|          |             | OR |
| ^        | XOR         |    |
| ~        | NOT         |    |
| <<       | Left shift  |    |
| >>       | Right shift |    |


```python
a = 5   # 101
b = 3   # 011

print(a & b)  # 1
print(a | b)  # 7
print(a ^ b)  # 6
```

8Ô∏è‚É£ Identity Operators

Check memory identity.

| Operator | Meaning          |
| -------- | ---------------- |
| is       | Same object      |
| is not   | Different object |


```python
a = [1,2]
b = a

print(a is b)   # True
```

```python
a = [1,2]
b = [1,2]

print(a == b)  # True
print(a is b)  # False
```

9Ô∏è‚É£ Membership Operators

Check presence.

| Operator | Meaning    |
| -------- | ---------- |
| in       | Exists     |
| not in   | Not exists |


```python
fruits = ["apple","banana"]

print("apple" in fruits)     # True
print("grape" not in fruits) # True
```

```python
"py" in "python"   # True
```
