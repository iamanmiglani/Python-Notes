What is a variable in Python?<br>
A variable is a name/reference that points to an object stored in memory.

Q: Stored by value or reference?<br>
By reference (object reference model).

Identity vs Equality

```python
a = [1,2,3]
b = [1,2,3]

print(a == b)   # True (values equal)
print(a is b)  # False (different objects)
```

Integer caching

```python
a = 256
b = 256
a is b   # True
```

```python
a = 1000
b = 1000
a is b   # Often False
```

#Python caches small integers (-5 → 256).

```python
x = [1,2,3]
y = x
del x
print(y)
print(x)

```

Shallow vs Deep Copy

Shallow copy

```python
import copy

a = [[1,2],[3,4]]
b = copy.copy(a)

a[0][0] = 99
print(b)
```

 Deep Copy

```python

a = [[1,2],[3,4]]
b = copy.deepcopy(a)
a[0][0] = 99
print(b)
```

```python
#df.copy(deep=True)   # Deep copy
#df.copy(deep=False)  # Shallow copy
```
