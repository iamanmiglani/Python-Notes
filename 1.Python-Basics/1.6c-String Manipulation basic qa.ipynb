```python
txt = "Miscommunication"
```

```python
# Reverse String
def reverse_string(s):
    return s[::-1]

reverse_string(txt)
```

```python
# Palindrome Check

def is_palindrome(s):
    return s == s[::-1]

is_palindrome(txt)
```

```python
# Reverse Words
txt = " I am DObby"

def reverse_words(s):
    return " ".join(s.split()[::-1])

reverse_words(txt)
```

```python
# Count Vowels & Consonants
txt = " I am in the most beautiful world"

def count_vc(s):
    v=set("aeiouAEIOU")
    vowels=sum(c in v for c in s)
    cons=sum(c.isalpha() and c not in v for c in s)
    return vowels, cons

count_vc(txt)
```

```python
# Remove Duplicates

def remove_dup(s):
    return "".join(dict.fromkeys(s))

remove_dup(txt)
```

```python
# First Non-Repeating Character

def first_unique(s):
    freq=Counter(s)
    for c in s:
        if freq[c]==1:
            return c

# without counter?
```
